```python
plt.figure(figsize=(12,6))
sns.histplot(data=data, x='num_subscribers', bins=100)
plt.xlim(0,40000)
plt.ylim(0,100)
plt.title("Number of Subscribers per Level")
plt.xlabel("Level")
plt.ylabel("No. of Subscribers")
plt.show()
```

# Group the data by subject and count the number of courses in each subject

```python
subject_counts =data.groupby("subject")["course_id"].count().reset_index(name="count")

# Sort the subjects by count in descending order
subject_counts = subject_counts.sort_values("count", ascending=False)

# Plot a bar chart of the subject counts
plt.figure(figsize=(10, 6))
plt.bar(subject_counts["subject"], subject_counts["count"], color="#377eb8")
plt.title("Number of Udemy courses by subject")
plt.xlabel("Subject")
plt.ylabel("Number of courses")
plt.xticks(rotation=90)
plt.show()
```

# showing all the courses which are free of cost ( using PIE CHART )

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the dataframe to include only the free courses
free_courses = data[data["is_paid"] == False]

# Count the number of courses in each subject
counts = free_courses["subject"].value_counts()

# Create a pie chart of the free courses by subject
plt.pie(counts, labels=counts.index, autopct="%1.1f%%")

# Set the title of the plot
plt.title("Free Udemy Courses by Subject")

# Show the plot
plt.show()
```

# Heat Map

```python
# Create a correlation matrix
corr_matrix = data.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, cmap='YlGnBu', annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
